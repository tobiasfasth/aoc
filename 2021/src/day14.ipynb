```python
from collections import defaultdict
import sys
import math

def parse_input(input_file):
    pairs = defaultdict(int)
    rules = defaultdict()
    
    f = open(input_file)
    template = f.readline().strip()
    input = f.read().strip().split('\n')

    for row in input:
        split = row.split('->')
        pair = split[0].strip()
        elem = split[1].strip()
        pairs[pair]
        rules[pair] = elem

    for c, elem in enumerate(template):
        if c < len(template)-1:
            pair = template[c]+template[c+1]
            pairs[pair] += 1

    return pairs, rules

def build_polymer(iterations, pairs, rules):
    for _ in range(iterations):
        diff = defaultdict(int)
        for pair, count in pairs.items():
            if count > 0:
                to_insert = rules[pair]
                first = pair[0]+to_insert
                second = to_insert+pair[1]
                diff[first] += count
                diff[second] += count
                diff[pair] -= count
        pairs = defaultdict(int,{key: pairs[key] + diff.get(key, 0) for key in pairs})
    
    return pairs

def get_element_count(polymer):
    elements = defaultdict(int)
    for key, value in polymer.items():
        elements[key[0]] += value
        elements[key[1]] += value

    min = sys.maxsize
    max = 0
    for key, value in elements.items():
        current_value = math.ceil(value/2)
        if  current_value > max:
            max = current_value
        elif current_value < min:
            min = current_value

    return max-min

# Part 1
input = '../data/day14.txt'
polymer = build_polymer(10, *parse_input(input))
nr_of_elements = get_element_count(polymer)
print(f"Part 1: {nr_of_elements}")

# Part 2
polymer = build_polymer(40, *parse_input(input))
nr_of_elements = get_element_count(polymer)
print(f"Part 2: {nr_of_elements}")
```
