```python
import copy

#part 2

f = open('../data/day04.txt', 'r')
drawn = list(map(int, f.readline().strip().split(",")))


def gen_matrix(str):
    lines = str.strip().splitlines()
    res = []
    for l in lines:
        res.append((list(map(int, l.split()))))
    return res


cards = list(map(gen_matrix, f.read().split("\n\n")))
cards_org = copy.deepcopy(cards)
complete_cards = [False for x in range(len(cards))]


def check_bingo(cards):
    true_check = False
    
    # check row
    for c, card in enumerate(cards):
        if complete_cards[c] == False:
            for r, row in enumerate(card):
                if (sum(row) == -5):
                    true_check = True
                    complete_cards[c] = True
                    return (True, c)

    #check column
    for c, card in enumerate(cards):
        if complete_cards[c] == False:
            for val in range(0,5):
                score = 0
                for row in card:
                    score += row[val]
                if score == -5:
                    true_check = True
                    complete_cards[c] = True
                    return (True, c)
    if not true_check:
        return (False, -1)


def bingo():
    last_card = -1
    last_drawn = -1
    for d in drawn:
        for c, card in enumerate(cards):
            if complete_cards[c] == False:
                for r, row in enumerate(card):
                    for v, value in enumerate(row):
                        if d == value:
                            cards[c][r][v] = -1
            if complete_cards[c] == False:
                t, w_c = check_bingo(cards)
                if t == True:
                    last_card = w_c
                    last_drawn = d

    return last_card, last_drawn


c, d = bingo()

unmarked_sum = 0

for row in cards[c]:
    for val in row:
        if val != -1:
            unmarked_sum += val

result = unmarked_sum*d
print(result)
```
