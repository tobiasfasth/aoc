```python
f = open('../data/day08.txt', 'r')


def split_line(line):
    line_split = line.split("|")
    signal = line_split[0].strip().split(" ")
    digit = line_split[1].strip().split(" ")
    return [signal, digit]


readings = list(map(split_line, f.read().strip().split('\n')))

count = 0
for l, line in enumerate(readings):
    for item in line[1]:
        if len(item) == 2 or len(item) == 3 or len(item) == 4 or len(item) == 7:
            count += 1

print('Part 1:', count)
```

```python
import re

total_output_values = []

for l, line in enumerate(readings):
        numbers = [0 for i in range(0,10)]

        for item in line[0]:
            if len(item) == 2:
                numbers[1] = item
            elif len(item) == 3:
                numbers[7] = item     
            elif len(item) == 4:
                numbers[4] = item
            elif len(item) == 7:
                numbers[8] = item

        x = 0
        while x < 6:
            for item in line[0]:
                if len(item) == 5: # 2,3,5
                    if len(re.findall('[' + str(numbers[7]) + ']', item)) == 3 and numbers[7] != 0:
                        numbers[3] = item
                        x += 1
                    elif len(re.findall('[' + str(numbers[9]) + ']', item)) == 5 and numbers[9] != 0:  
                        numbers[5] = item
                        x += 1
                    elif len(re.findall('[' + str(numbers[9]) + ']', item)) == 4 and numbers[9] != 0:
                        numbers[2] = item
                        x += 1
                elif len(item) == 6: # 0,6,9
                    if item == 'cbegaf':
                        print(numbers)
                    if len(re.findall('[' + str(numbers[7]) + ']', item)) == 2 and numbers[7] != 0:
                        numbers[6] = item
                        x += 1
                    elif len(re.findall('[' + str(numbers[4]) + ']', item)) == 4 and numbers[4] != 0:
                        numbers[9] = item
                        x += 1                        
                    elif len(re.findall('[' + str(numbers[8]) + ']', item)) == 6 and numbers[8] != 0:
                        numbers[0] = item
                        x += 1      

        output_value = ""
        for item in line[1]:
            if bool(re.search('^[' + str(numbers[0]) + ']{6}$', item)):
                output_value += '0'
            elif bool(re.search('^[' + str(numbers[1]) + ']{2}$', item)):
                output_value += '1'
            elif bool(re.search('^[' + str(numbers[2]) + ']{5}$', item)):
                output_value += '2'
            elif bool(re.search('^[' + str(numbers[3]) + ']{5}$', item)):
                output_value += '3'
            elif bool(re.search('^[' + str(numbers[4]) + ']{4}$', item)):
                output_value += '4'
            elif bool(re.search('^[' + str(numbers[5]) + ']{5}$', item)):
                output_value += '5'
            elif bool(re.search('^[' + str(numbers[6]) + ']{6}$', item)):
                output_value += '6'
            elif bool(re.search('^[' + str(numbers[7]) + ']{3}$', item)):
                output_value += '7'
            elif bool(re.search('^[' + str(numbers[8]) + ']{7}$', item)):
                output_value += '8'
            elif bool(re.search('^[' + str(numbers[9]) + ']{6}$', item)):
                output_value += '9'

        total_output_values.append(int(output_value))

print('Part 2:', sum(total_output_values))
```
